```python
import bs4 as bs
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
in_url='https://www.amazon.in/s?k=fruit+juice&crid=2QAJKUKCEF3N5&sprefix=fruit+%2Caps%2C266&ref=nb_sb_ss_ts-doa-p_4_6'
headers={
    'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36',
    'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
    'Accept-Language':'en-US,en;q=0.5',
    'Acept-Encoding':'gzip',
    'DNT':'1',
    'Connection':'closed'}
res=requests.get(in_url,headers=headers)
soup=BeautifulSoup(res.text,'html.parser')    
```

```python
items=soup.find_all('div',class_="puis") # class you need to search in html
print(items[0].prettify()) # this will give you idea about the html code of site under inspect
```

```python
article_content = []
for item in items:
  item_name=item.find('span',class_="a-size-base-plus a-color-base a-text-normal")
  item_price=item.find('span',class_="a-price-whole")
  item_code=item.find('span',class_="a-size-base s-underline-text")
  item_discount=item.find('span', class_="a-size-base a-color-secondary")

  if item_price is None or item_name is None:
    print(':'+'Price not Available')
  else:
    print(item_name.text+':'+item_code.text+':'+item_price.text+':'+item_discount.text)
  
  article_content.append({
                          "name": item_name.text if item_name is not None else '',
                          "item_price":item_price.text if item_price is not None else '',
                          "item_code":item_code.text if item_code is not None else '',
                          "item_discount":item_discount.text if item_discount is  not None else ''})

print(article_content)
```

```python
df = pd.DataFrame(article_content)
df.to_csv('article_content.csv', index=False)
```
