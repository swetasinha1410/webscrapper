**Import Libraries**

```python
import numpy as np
import pandas as pd
import urllib
import nltk
import bs4 as bs
import re
#import nltk.corpus import stopwords
nltk.download('stopwords')
```

**Load Data**

```python
source=urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming').read()
```

**Prasing the data and creating BeautifulSoup**

```python
soup=bs.BeautifulSoup(source,'lxml')
```

Featching the data

```python
text=""
for paragraph in soup.find_all('p'):
  text+=paragraph.text
```

```python
text
```

```python
text=[text]
text
```

**Preprocessing of data**

```python
def preprocessText(x):
    temp = x.lower()
    temp = re.sub(r'[[0-9]*\]', ' ', temp)
    temp = re.sub(r'[^\w]', ' ', temp)#regex : pattern based searching of substring in the string
    #  ^ : NOT \w: for every lower case character/word
    #if a character is not a lower case word replace it with a space
    return temp


```

```python
articles_final = [preprocessText(text) for text in text]
```
