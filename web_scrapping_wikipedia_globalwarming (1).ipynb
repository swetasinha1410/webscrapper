**Preprocessing of data**

```python
import nltk
from gensim import corpora
from gensim.corpora import Dictionary
from nltk.stem import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
import re
from nltk.corpus import stopwords

import glob
import os

# For displaying images in ipython
from IPython.display import HTML, display 
import pprint
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
import nltk
nltk.download('stopwords')
     
```

```python
import nltk
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
def preprocessText(x):
    temp = x.lower()
    temp = re.sub(r'[[0-9]*\]', ' ', temp)
    temp=re.sub(r'\s+',"",temp) # removing extra space
    temp = re.sub(r'[^\w]', ' ', temp)#regex : pattern based searching of substring in the string
    #  ^ : NOT \w: for every lower case character/word
    #if a character is not a lower case word replace it with a space

    temp = nltk.word_tokenize(temp)
    temp = [wordnet_lemmatizer.lemmatize(w) for w in temp] # refind version of steeming, will remove true ing from root word
    temp = [word for word in temp if word not in stopwords ]
    return temp
    return temp
  
    
```

```python

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
articles_final = [preprocessText(text) for text in text]
```

```python
articles_final
```

**Transformation of Preprocessed text into Vector form using Gensim**



```python

# Create a dictionary representation of the documents.
dictionary = Dictionary(articles_final)

# Filter out words that occur less than 20 documents, or more than 50% of the documents.
#dictionary.filter_extremes(no_below=20, no_above=0.5)
     
```

```python
print(dictionary)
     
```

```python
# Bag-of-words representation of the documents.
corpus = [dictionary.doc2bow(doc) for doc in articles_final]
```

```python

print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
     
```

```python

#Train LDA Model
from gensim.models import LdaModel
#setting training parameters
num_topics=5
chunksize=200
passes=20
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token
# print(id2word)

model=LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    #iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every)
```
